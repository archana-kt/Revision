## Subplot

```python
plt.subplot(3,1,1) # 3 rows 1 column and 1st set
plt.plot(x,np.sin(x))

plt.subplot(3,1,2) # 3 rows 1 column and 2nd set
plt.plot(x,np.cos(x))

plt.subplot(3,1,3) # 3 rows 1 column and 3rd set
plt.plot(x,np.sin(x))
```

```python
# Another method, instead of subplot
```

```python
fig, ax = plt.subplots(2)  # Use plt.subplots to create a figure and subplots
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## legend

```python
x
```

```python
plt.plot(x,np.sin(x),label="sin")
```

```python
plt.plot(x,np.cos(x),label="cos")
```

```python
# See we are not getting legend
```

```python
# To get that, we need to declare
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') # green
plt.plot(x, np.cos(x), ':b', label='cos(x)') # blue
plt.axis('equal')
plt.legend()
```
